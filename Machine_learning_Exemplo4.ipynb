```
import pandas as pd
url = "https://raw.githubusercontent.com/alura-cursos/machine-learning-algoritmos-nao-supervisionados/master/movies.csv"
filmes = pd.read_csv(url)

troca = {
    "movieId" : "id",
    "title" : "titulo",
    "genres" : "generos"
}

filmes = filmes.rename(columns = troca)
```

**Identificando dummies e concatenando dataframes**

```
# axis 0 = um embaixo do outro, axis 1 = lado a lado
filmes = pd.concat([filmes, filmes.generos.str.get_dummies()], axis = 1)
```

**Nota**: *Não é possível escalar (scaler) os dados binários pois não são atributos preparados para isso. O exemplo é apenas para efeitos didáticos*

```
# isola matriz de gêneros
generos_filmes = filmes.generos.str.get_dummies()
generos_filmes
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
generos_escalados = scaler.fit_transform(generos_filmes)
```
